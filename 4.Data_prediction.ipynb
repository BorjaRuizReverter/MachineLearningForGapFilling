```python
import joblib

regreModel = joblib.load('regreModel.model')
```

```python
import pandas as pd

X_december = pd.read_pickle('X_later.pkl')
y_december = pd.read_pickle('y_later.pkl')
```

```python
y_december_prediction = regreModel.predict(X_december)
y_december_prediction
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.plot(y_december['NEE'].to_numpy(), color = 'red', marker = '.', linestyle = '-', label = 'Real data')
plt.plot(y_december_prediction,        color = 'blue', marker = '.', linestyle = ' ', label = 'Predicted data')
plt.xlabel('Time from Dezember2013 (30min)', fontsize=18)
plt.ylabel('NEE ($\mu$mol m$^{-2}$ s$^{-1}$)', fontsize=18)
plt.xlim(0,1350)
plt.legend()

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
plt.tight_layout()

# When saving in png the captions are capped. None of the lines above helped, so I change to jpg.
#plt.savefig('assets/output.png', bbox_inches="tight")
plt.savefig('assets/output.jpg', bbox_inches="tight")
```
